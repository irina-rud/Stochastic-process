Напишем класс, будет моделировать процесс описанный в задаче 3.6. В задаче $t \in [0, \pi]$, но нам нужно моделировать процесс в том числе и для $t > \pi$, чтобы корректно продлить процесс, будем склеивать его в точках кратных $\pi$.

Для каждого промежутка кратного $\pi$, eсли нужно, то генерируем свои случайные величины и запоминаем их в map. 

```python
from random import normalvariate
from random import expovariate

import math
from matplotlib import mlab

import pylab
%pylab inline



class WinerProcess:
    def __init__(self, n):
        self.accuracy = n
        self.xi = {}
    def gen(self, start):
        xiTmp = []
        for i in range(self.accuracy + 1):
            xiTmp.append(normalvariate(0, 1))
        self.xi[start] = xiTmp

    def calc(self, time):
        number = int(time // math.pi)
        if(not (number in self.xi)):
            self.gen(number)
        answer = self.xi[number][0] * (time % math.pi) / (math.pi)**(0.5)
        for i in range(self.accuracy):
            answer += (2.0 / math.pi) ** (0.5) * (math.sin((i + 1) * time % math.pi)/(i + 1)) * self.xi[number][i + 1]
        if(number > 0):
            delta = 0.0
            for i in range(number):
                if(not(i in self.xi)):
                    self.gen(i)
                delta += self.xi[i][0] * math.pi / (math.pi **0.5)
            answer += delta
        return answer
    def getitem(self, times):
        return list(map(self.calc, times))
    def plot(self, step, a, b):
        xlist = mlab.frange (a, b, step)
        ylist = self.getitem(xlist)
        pylab.plot(ylist)
        pylab.show()
```

Построим 3 графика меняя n от 10 до 1000.

```python
winer10 = WinerProcess(10)
winer10.plot(1/1000, 0, 1)
winer100 = WinerProcess(100)
winer100.plot(1/1000, 0, 1)
winer1000 = WinerProcess(1000)
winer1000.plot(1/1000, 0, 1)
```

Видно, что при увеличении n график начинает колебаться больше.

Построим графики уменьшая мастаб, но сохраняя количество точек на графике постоянным.

```python
winer1000.plot(1/10, 0, 100)
winer1000.plot(1/100, 0, 10)
winer1000.plot(1/1000, 0, 1)
```

Траектории, как и должно было получиться остаются неразличимыми.

Будем увеличивать масштаб. Не изменяя n. Для процесса, где n не очень большое.

```python
winer100.plot(1/100, 0, 10)
winer100.plot(1/100, 0, 1)
```

Видно, что когда точек на графике становится мало, процесс выглядит как склейка линейных функций. Хотя на самом деле для винеросовского процесса, это ни при каком увеличении масштаба не верно.
```python
class modelMO:
    def __init__(self, ymax, ymed, ymin):
        self.a = []
        self.b = []
        self.a.append(normalvariate(0, 1))
        self.b += [normalvariate(0, 1)]
        self.timeNextStep = []
        self.timeNextStep.append(0)
        self.startBalance = []
        self.startBalance.append(ymed)
        self.win = WinerProcess(100)
        self.ymax = ymax
        self.ymed = ymed
        self.ymin = ymin
        #print("end init")
    
    def countAnswer(self, time, number):
        answer = 0
        answer = self.startBalance[number - 1]
        answer += self.a[number -1] * (time - self.timeNextStep[number - 1])
        answer += ((time - self.timeNextStep[number - 1])* self.b[number - 1] * time)
    #    print(time)
        answer += self.win.getitem([time])
        return answer
    
    def calc(self, time):
        step = len(self.timeNextStep)
        if(self.timeNextStep[step - 1] <= time):
            while(time >= self.timeNextStep[step - 1]):
                exp = expovariate(0.5)
                self.timeNextStep.append(self.timeNextStep[step - 1] + exp)
                self.a.append(normalvariate(0, 1))
                self.b.append(normalvariate(0, 1))
                self.startBalance.append(self.startBalance[step - 1] + self.a[step] * exp + 
                                         self.b[step] * (exp) * self.timeNextStep[step])
                step += 1
     #   print(self.timeNextStep)
        number = 0
        while(self.timeNextStep[number] <= time):
            number += 1
      #  print(number)
        answer = self.countAnswer(time, number)
        
        if(answer[0] <= self.ymin):
            self.a[number - 1] = normalvariate((self.ymed - self.ymin), 1)
            self.b[number - 1] = normalvariate(1, 1)
            answer = self.countAnswer(time, number)
        elif(answer[0] >= self.ymax):
            self.a[number - 1] = expovariate((-(self.ymed - self.ymin))/ 2)
            self.b[number - 1] = normalvariate(-1, 1)
            answer = self.countAnswer(time, number)
        return answer
    def getitem(self, times):
        a = []
        for t in times:
            b = self.calc(t)
            a += b.tolist()
        return a
        #return list(map(self.calc, times))
    def plot(self, step, a, b):
        xlist = mlab.frange (a, b, step)
        ylist = self.getitem(xlist)
      #  print(ylist)
        pylab.plot(ylist)
        pylab.show()
```
